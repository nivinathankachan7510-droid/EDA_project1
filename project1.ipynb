Life Style Data

Introduction

This dataset captures detailed information about individual health, body measurements, workout characteristics, and calorie expenditure.
Key variables include age, gender, weight, height, types of workout, session duration, heart-rate data (resting, average, and max BPM), calories burned, and additional nutritional and metabolic features such as macro-based calorie intake, lean mass, heart-rate reserve percentage, and calorie balance.

The goal of this dataset is to explore how different physiological attributes and exercise patterns influence calorie burn and fitness performance.
The data provides a unique opportunity to analyze which workout types are most effective, how body composition affects calorie expenditure, and how heart rate metrics correlate with physical effort.


The dataset includes:

Personal & Demographic Information

* Age

* Gender

* Weight (kg)

* Height (m)

* BMI, BMI_calc

* Lean mass (kg)

Heart-Rate & Fitness Metrics

* Max BPM

* Avg BPM

* Resting BPM

* % Heart-Rate Reserve (pct_HRR)

* % of Maximum HR (pct_maxHR)

* Workout Frequency (days/week)

* Session Duration (hours)

* Expected calorie burn

* Burn category (Burns_Calories_Bin)

 Workout & Exercise Characteristics

* Workout Type

* Name of Exercise

* Sets & Reps

* Target Muscle Group

* Body Part

* Equipment Needed

* Benefit description

* Physical exercise

* Burns Calories (per 30 min)

 Nutrition & Food Information

* Daily meals frequency

* Meal name, meal type, diet type

* Cooking method

* Serving size (g)

* Calories & macro nutrients (Carbs, Proteins, Fats)

* Macro calorie calculation (cal_from_macros)

* Protein per kg body weight

* Calorie balance (cal_balance)

 Food Composition (Micro-Nutrient Values)

2.How do Max BPM and Average BPM differ across various workout types?

```python
# Group by workout type and calculate mean BPM values
bpm_compare = df.groupby('Workout_Type')[['Max_BPM','Avg_BPM']].mean().sort_values(by='Max_BPM', ascending=False)

print("Max BPM & Average BPM across Workout Types:\n")
print(bpm_compare)
```

3.Which workout type burns the most calories per hour?

```python
# Calculate Calories Burned Per Hour
df['Calories_per_Hour'] = df['Calories'] / (df['Session_Duration (hours)'] / 60)

# Group by workout type
cal_per_hour = df.groupby('Workout_Type')['Calories_per_Hour'].mean().sort_values(ascending=False)

print("Average Calories Burned Per Hour by Workout Type:\n")
print(cal_per_hour)



```

4.Which age group appears to have the highest fitness performance (based on BPM or calories burned)?

```python
age_fitness = df.groupby('Age')[['Avg_BPM', 'Calories']].mean().sort_values(by='Calories', ascending=False)

print("Fitness Performance by Age Group:")
print(age_fitness)



```

5.Classify users into categories (High/Medium/Low calorie burners)?

```python
# Classify users based on Calories Burned
df['Calorie_Category'] = pd.qcut(df['Calories'],q=3,labels=['Low Burner', 'Medium Burner', 'High Burner'])

# Show first few rows
print(df[['Calories', 'Calorie_Category']].head())

# Count of each category
print("\nCategory Counts:")
print(df['Calorie_Category'].value_counts())
```

7.Which Exercise helps to  reduce fat the most?

```python


exercise_effectiveness = df.groupby("Name of Exercise")["Fat_Percentage"].count().sort_values(ascending=False)

print("\nFat reduction by diet type:")
print(exercise_effectiveness)

# Visualization
plt.figure(figsize=(10,5))
exercise_effectiveness.plot(kind='bar')
plt.title("Average Fat Loss by Exercise")
plt.xlabel("Exercise Name")
plt.ylabel("Average Fat Loss (Fat Percentage Reduced)")
plt.tight_layout()
plt.show()
```

8.Which muscle is most developed when you workout with which equipment?

```python
muscle_development = (df.groupby(["Equipment Needed", "Workout"])["Type of Muscle"].count().reset_index().sort_values(by="Type of Muscle", ascending=False))

print("\nMost developed muscles by equipment:")
print(muscle_development)

# Visualization
plt.figure(figsize=(14,6))
sns.barplot(data=muscle_development, x="Equipment Needed", y="Type of Muscle")
plt.title("Muscle Development by Equipment ")
plt.xlabel("Equipment ")
plt.ylabel("Muscle Growth")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

9.Which cooking method gives the best protein and calories?

```python


cooking_method_analysis = (
    df.groupby("cooking_method")[["Proteins", "Calories"]]
      .mean()
      .sort_values(by=["Proteins", "Calories"], ascending=False)
)

print("\nAverage Protein & Calories by Cooking Method:")
print(cooking_method_analysis)

plt.figure(figsize=(12,6))
sns.boxplot(x="cooking_method", y="Proteins", data=df)
plt.title("Protein Distribution by Cooking Method")
plt.xlabel("Cooking Method")
plt.ylabel("Protein Content")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12,6))
sns.boxplot(x="cooking_method", y="Calories", data=df)
plt.title("Calories Distribution by Cooking Method")
plt.xlabel("Cooking Method")
plt.ylabel("Calories")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

10.Which cooking method is good based on prep_time_min and cook_time_min?
